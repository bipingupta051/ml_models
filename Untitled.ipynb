```python
def naive_bayes(df,target_col,instances):
    
    class_probability=df[target_col].value_counts()/len(df)
    max_probability=-1
    predicted_class=None
    for class_value in class_probability.index:
        probability=class_probability[class_value]
        for attribute,attribute_value in instances.items():
            filterd_data=df[df[target_col]==class_value]
            attribute_counts=filterd_data[attribute].value_counts().get(attribute_value,0)
            class_count=len(filterd_data)
            probability *=attribute_counts/class_count
        print(f" probability of '{class_value}':{probability}")
        if (probability>max_probability):
            
            max_probability=probability
            predicted_class=class_value
    return predicted_class
                
        
                
                
        
    
```

```python
new_instances={'age':'youth','income':'medium','student':'yes','credit_rating':'fair'}
class_label=input("Enetr the class label name")
predicted_class=naive_bayes(df,class_label,new_instances)
print(f" predicted class for given instances: ", predicted_class)
```

```python
def BV(df): #conversion into boolean vector
    items=dict()
    for i in range(len(df)):
        data=df['Items'][i].split(',')
        for item in data:
            items[item]=list()
    for i in range(len(df)):
        data=df['Items'][i].split(',')
        for key in items.keys():
            if key in data:
                items[key].append(1)
            else :
                items[key].append(0)
    items["TID"]=range(1,len(df)+1)
    return items
            
```
